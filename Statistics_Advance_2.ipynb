1.  What are Type I and Type II errors in hypothesis testing, and how do they
impact decision-making?

    ANS:- In hypothesis testing, two types of errors can occur. A Type I error happens when the null hypothesis is true but is incorrectly rejected leading to a false positive conclusion. A Type II error occurs when the null hypothesis is false but fails to be rejected resulting in a false negative. These errors directly affect decision making: Type I errors may cause acceptance of ineffective treatments or policies, while Type II errors may overlook beneficial interventions. Balancing the risks of both errors is crucial, often achieved by adjusting significance levels and sample sizes to minimize their impact on conclusions.


2. What is the P-value in hypothesis testing, and how should it be interpreted
in the context of the null hypothesis?

    ANS:- The `p-value` in hypothesis testing represents the probability of obtaining results as extreme as those observed assuming the null hypothesis is true. It serves as a measure of evidence against the null hypothesis. A `small p-value` (commonly ≤ 0.05) suggests that the observed data are unlikely under the null leading to rejection of the null hypothesis in favor of the alternative. A `large p-value` indicates that the data are consistent with the null so it is not rejected. Importantly, the `p-value` does not show the probability that the null hypothesis is true it only reflects data compatibility.


3. Explain the difference between a Z-test and a T-test, including when to use
each.

    ANS:- A `Z-test` and a `T-test` are statistical methods used to compare sample means with population means or between groups. The `Z-test` relies on the standard normal distribution and is appropriate when the population variance is known and the sample size is large (n ≥ 30). In contrast, the `T-test` uses the Student's `t-distribution` which accounts for greater uncertainty, and is applied when the population variance is unknown and the sample size is small (n<30). As sample size increases, the `t-distribution` approaches the normal distribution making both tests yield similar results in large sample scenarios.

4. :What is a confidence interval, and how does the margin of error influence
its width and interpretation?

    ANS:- A confidence interval is a statistical range around a sample estimate that is likely to contain the true population parameter at a chosen confidence level such as 95%. It is expressed as the estimate plus or minus the margin of error. The margin of error depends on sample size, variability, and confidence level, and it directly controls the interval's width. A larger margin produces a wider interval, reflecting greater uncertainty, while a smaller margin yields a narrower interval, indicating more precision. Interpreting a 95% confidence interval means that repeated sampling would capture the true parameter in 95% of cases.

5. Describe the purpose and assumptions of an ANOVA test. How does it
extend hypothesis testing to more than two groups?

    ANS:- ANOVA, or Analysis of Variance is a statistical method used to test whether the means of three or more groups differ significantly. Its purpose is to avoid multiple `t-tests`, which increase the risk of `Type I errors` by providing a single test for overall group differences. The assumptions of ANOVA include independence of observations, normally distributed data within groups, and equal variances across groups. It extends hypothesis testing beyond two groups by comparing variance between groups to variance within groups. If between-group variance is significantly larger, the null hypothesis of equal means is rejected, indicating meaningful differences.


6.  Write a Python program to perform a one-sample Z-test and interpret the
result for a given dataset.

```python
import numpy as np
from scipy.stats import norm

data = [52, 55, 53, 50, 51, 54, 56, 52, 53, 55]
sample_mean = np.mean(data)
population_mean = 50
population_std = 3
n = len(data)

z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

p_value = 2 * (1 - norm.cdf(abs(z_score)))

print("Sample Mean:", sample_mean)
print("Z-Score:", z_score)
print("P-Value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: sample mean differs significantly.")
else:
    print("Fail to reject the null hypothesis: no significant difference.")
```

8. Generate multiple samples from a non-normal distribution and implement
the Central Limit Theorem using Python.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
scale = 2.0
num_samples = 1000
sample_size = 50

sample_means = []
for _ in range(num_samples):
    sample = np.random.exponential(scale=scale, size=sample_size)
    sample_means.append(np.mean(sample))

plt.hist(sample_means, bins=30, density=True, color='mediumseagreen', edgecolor='black')
plt.title("Central Limit Theorem Demonstration\nSample Means from Exponential Distribution")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.show()
```

9.  Write a Python function to calculate and visualize the confidence interval
for a sample mean.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def calculate_confidence_interval(data, confidence=0.95):
    """
    Calculate and visualize the confidence interval for a sample mean.

    Parameters:
    - data: array-like, the sample data
    - confidence: float, confidence level (e.g., 0.95 for 95%)

    Returns:
    - mean: sample mean
    - lower_bound: lower bound of the confidence interval
    - upper_bound: upper bound of the confidence interval
    """
    n = len(data)
    mean = np.mean(data)
    sem = stats.sem(data)
    margin = sem * stats.t.ppf((1 + confidence) / 2., n-1)
    lower_bound = mean - margin
    upper_bound = mean + margin

    plt.style.use('seaborn-v0_8')
    plt.figure(figsize=(8, 5))
    plt.hist(data, bins=20, color='cornflowerblue', edgecolor='black', alpha=0.7, density=True)
    plt.axvline(mean, color='red', linestyle='--', label=f'Mean = {mean:.2f}')
    plt.axvline(lower_bound, color='green', linestyle='--', label=f'Lower CI = {lower_bound:.2f}')
    plt.axvline(upper_bound, color='green', linestyle='--', label=f'Upper CI = {upper_bound:.2f}')
    plt.title(f'Confidence Interval ({int(confidence*100)}%) for Sample Mean')
    plt.xlabel('Sample Values')
    plt.ylabel('Density')
    plt.legend()
    plt.show()

    return mean, lower_bound, upper_bound

np.random.seed(0)
sample_data = np.random.exponential(scale=2.0, size=100)
calculate_confidence_interval(sample_data, confidence=0.95)
```

10.  Perform a Chi-square goodness-of-fit test using Python to compare
observed and expected distributions, and explain the outcome.

```python
"""
Chi-square Goodness-of-Fit Test
This script compares an observed distribution to an expected distribution
and explains the outcome.
"""

import numpy as np
from scipy.stats import chisquare


# Observed and Expected Values

# Example observed counts across 4 categories

observed = np.array([50, 30, 15, 5])

# Expected distribution (equal probability for each category)

expected_probs = np.array([0.25, 0.25, 0.25, 0.25])
expected = expected_probs * observed.sum()


# Perform Chi-square Test

chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square Statistic:", chi2_stat)
print("p-value:", p_value)


# Interpretation

alpha = 0.05
if p_value < alpha:
    conclusion = (
        "Since the p-value is less than 0.05, we reject the null hypothesis.\n"
        "The observed distribution differs significantly from the expected distribution."
    )
else:
    conclusion = (
        "Since the p-value is greater than 0.05, we fail to reject the null hypothesis.\n"
        "The observed distribution does NOT differ significantly from the expected distribution."
    )

print("\nInterpretation:")
print(conclusion)
```
